# Ejercicio de composición de funciones

Un estudio ambiental de cierta comunidad suburbana sugiere que el nivel diario promedio de monóxido de carbono en el aire será 

$$ C(p) = 0.5p+3 $$

partes por millón cuando la población sea de $p$ miles.   

Se estima que en $t$ años la población de la comunidad será 

$$ p(t) = 6 + 0.3t^{2} \text{miles} $$


1. Exprese el nivel de monóxido de carbono en el aire como una función del tiempo.
2. ¿Cuál será el nivel de monóxido de carbono en 4 años, a partir de hoy?
3. ¿Cuándo alcanzará el nivel de monóxido de carbono de 9 partes por millón?

## Solución:

1. 

$$ C(t) = 0.5(6+0.3t^{2}) + 3 = 3 + \frac{3}{20}t^{2} + 3 = 6 + \frac{3}{20}t^{2} $$

## Graficación de la composición

```python
import numpy as np
import matplotlib.pyplot as plt 

fig = plt.figure(figsize = (14,10))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)


t = np.linspace(0,5)
C = 6 + (3/20)*t**2
p = 6 + 0.3*t**2

ax1.plot(p,C)
ax1.set_title('$C(p)$')
ax1.set_xlabel('p')
ax1.set_ylabel('C')
ax1.grid(True)


ax2.plot(t,p)
ax2.set_title('$p(t)$')
ax2.set_xlabel('t')
ax2.set_ylabel('p')
ax2.grid(True)

ax3.plot(t,C)
ax3.set_title('$C(t)$')
ax3.set_xlabel('t')
ax3.set_ylabel('C')
ax3.grid(True)

plt.savefig('composicion.jpg')

```
