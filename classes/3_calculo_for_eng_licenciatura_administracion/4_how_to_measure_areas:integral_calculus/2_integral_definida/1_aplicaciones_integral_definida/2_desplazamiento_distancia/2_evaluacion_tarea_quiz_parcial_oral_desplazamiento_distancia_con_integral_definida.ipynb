```python
# valoracion de la asistencia  

import pandas as pd
import numpy as np

# Cargar la lista de estudiantes desde la URL
path = '/home/marco/Documentos/docencia/groups_list/calculo_vectorial.csv'
estudiantes = pd.read_csv(path)
estudiantes['Parcial'] = estudiantes['Parcial'].astype(float) # hacer que todos los valores de la columna Parcial seen de tipo float

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def asistencia():
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Asistió {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        if asistencia == 's':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Parcial'] += 0.1
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Parcial']} en la columna 'Parcial'.")
        
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Guardar los cambios en el archivo CSV
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
while lista_estudiantes:
    asistencia()
    guardar_cambios()
```

```python
# Tarea 1 de lectura y completación de ejemplos 
# asignación de tarea para realizar antes de clase
# Consiste en leer y completar un ejemplo de texto guía

import pandas as pd
import numpy as np
import requests
np.random.seed(1213)  # establecer semilla aleatoria para hacer los resultados reproducibles



grupo_estudiantes = 'calculo_vectorial'
nombre_texto_guia = 'Purcell'
pagina_y_seccion = 'Página_275_Sección_5.1'
numero_ejemplos_de_la_seccion = 7
numero_de_ejemplos_para_leer_y_completar = 6


# Descargar la lista de  estudiantes
students_path = "/home/marco/Documentos/docencia/groups_list/" + grupo_estudiantes + ".csv"
exercises = list(range(1, numero_ejemplos_de_la_seccion))  

students_data = pd.read_csv(students_path)

# Crear una lista para almacenar las asignaciones
assignments = []

# Asignar ejercicios aleatorios a cada estudiante
for index, student in students_data.iterrows():
    # Seleccionar 5 ejercicios aleatorios sin repetición
    random_exercises = np.sort(np.random.choice(exercises, numero_de_ejemplos_para_leer_y_completar, \
                                        replace = False) + 1 )
    assignments.append([students_data["Nombre"][index], *random_exercises])  # Agregar asignación a la lista

# Crear un DataFrame con las asignaciones
columns = ["Student"] + [ "Example" + str(i+1) for i in range(numero_de_ejemplos_para_leer_y_completar)]
assignment_df = pd.DataFrame(assignments, columns=columns)

# Guardar el DataFrame en un archivo CSV
assignment_csv = "tarea" + "_" + nombre_texto_guia + "_"+ pagina_y_seccion + ".csv"
assignment_df.to_csv(assignment_csv, index=False)

print("Tarea 1 sobre lectura a ejemplos: ", assignment_csv)
assignment_df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Metodología
# 1. Graficar la región
# 2. Plantear la integral que define al área de la región
# 3. Evaluar la integral  

plt.axis([-3, 4, -5, 5])
incremento = 0.1
dominio = np.arange(-2, 3 + incremento, incremento)

def f(x):
    return -(x**2 / 3 - 4)

rango = f(dominio)
plt.plot(dominio, rango, color='red')
plt.fill_between(dominio, rango, color = 'blue')
# Agregar la línea horizontal en y = 0
plt.axhline(y=0, color='green')

# Agregar las líneas verticales en x = -2 y x = 3
plt.axvline(x=-2, color='green')
plt.axvline(x=3, color='green')

plt.show()
```

32. Un objeto se mueve a lo largo de una recta, de modo que su velocidad en el instante $t$ es $v(t) = \frac{1}{2} + sen (2t)$ pies por segundo. Encuentre el desplazamiento y la distancia recorrida por el objeto en el intervalo $0 \leq t \leq \frac{3\pi}{2}$.

```python
import matplotlib.pyplot as plt 
import numpy  

dominio = np.arange(0,3*np.pi/2, 0.1)

def v(t):
    return 1/2 + np.sin(2*t)
rango = v(dominio) 
plt.plot(dominio, rango) 

plt.axhline(y = 0, color = 'green') 
```

# Programa para evaluación del seguimiento mensual de los estudiantes

```python
# Programa para valoración cuantitativa por meses para los estudiantes de todos los cursos
import numpy as np 
import pandas as pd   

#weights
weightAssistance = 12/100
weightAttentionParticipation = 8/100
weightHandwrittenNotebook = 8/100
weightJupyterNotebook = 20/100
weightQuiz = 16/100
weightPartial = 48/100

url_group = 'https://raw.githubusercontent.com/marco-canas/groups_list/main/seminario_didactica_aritmetica.csv'

df = pd.read_csv(url_group, decimal = ',') 

df.Assistance = (df.Assistance/np.max(df.Assistance))*5.0
df.AttentionParticipation = (df.AttentionParticipation/np.max(df.AttentionParticipation))*5.0
df.HandwrittenNotebook = (df.HandwrittenNotebook/np.max(df.HandwrittenNotebook))*5.0
df.JupyterNotebook = (df.JupyterNotebook/np.max(df.JupyterNotebook))*5.0  


df['FollowUp'] = df.Assistance*weightAssistance + \
                 df.AttentionParticipation*weightAttentionParticipation + \
                 df.HandwrittenNotebook*weightHandwrittenNotebook+\
                 df.JupyterNotebook*weightJupyterNotebook+\
                 df.Quiz*weightQuiz + \
                 df.Partial*weightPartial

df.FollowUp = df.FollowUp.round(1)

df[['Nombre', 'FollowUp']] 
```
