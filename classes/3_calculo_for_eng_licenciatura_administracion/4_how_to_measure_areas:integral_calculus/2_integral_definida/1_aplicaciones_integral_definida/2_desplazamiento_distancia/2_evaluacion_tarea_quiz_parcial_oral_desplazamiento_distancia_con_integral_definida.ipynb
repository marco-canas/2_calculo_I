```python
# valoracion de la asistencia  

import pandas as pd
import numpy as np

# Cargar la lista de estudiantes desde la URL
path = '/home/marco/Documentos/docencia/groups_list/calculo_vectorial.csv'
estudiantes = pd.read_csv(path)
estudiantes['Parcial'] = estudiantes['Parcial'].astype(float) # hacer que todos los valores de la columna Parcial seen de tipo float

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def asistencia():
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Asistió {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        if asistencia == 's':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Parcial'] += 0.1
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Parcial']} en la columna 'Parcial'.")
        
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Guardar los cambios en el archivo CSV
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
while lista_estudiantes:
    asistencia()
    guardar_cambios()
```

# Ejercicios 32 de la sección 5.1 del Purcell página 280 del PDF y del libro  


32. Un objeto se mueve a lo largo de una recta, de modo que su velocidad en el instante $t$ es $v(t) = \frac{1}{2} + sen (2t)$ pies por segundo. Encuentre el desplazamiento y la distancia recorrida por el objeto en el intervalo $0 \leq t \leq \frac{3\pi}{2}$.

```python
import matplotlib.pyplot as plt 
import numpy  
a,b = 0, 3*np.pi/2
dominio = np.arange(0,3*np.pi/2, 0.1)

def v(t):
    return 1/2 + np.sin(2*t)

rango = v(dominio) 
plt.figure(figsize = (8,4))
plt.title('Gráfico de Velocidad vs tiempo') 
plt.xlabel('Tiempo t')
plt.ylabel('Velocidad v')
plt.plot(dominio, rango, color = 'red', label = 'Velocidad en cada instante') 
plt.fill_between(dominio, rango, color = 'blue')
plt.axvline(x = 0, ymin = 0, ymax = v(0), color = 'green')
plt.axvline(x = 3*np.pi/2, ymin = 0, ymax = v(3*np.pi/2) ,color = 'green') 
plt.axhline(y = 0, color = 'green') 
plt.grid(alpha = 0.3)
plt.xticks(np.arange(a,b, 0.2), rotation = 90) # metrizar el eje horizontal

plt.savefig('/home/marco/Documentos/docencia/calculo/images/4_integral/2_aplicaciones_integral_definida/3_desplazamiento_distancia/velocidad_senoidal_ejeccicio_32.png')
```
