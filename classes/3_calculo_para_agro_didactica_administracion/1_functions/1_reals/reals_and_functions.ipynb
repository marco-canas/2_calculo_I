# Mapa conceptual de clasificación de los números reales en forma de arbol

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un gráfico dirigido
G = nx.DiGraph()

# Agregar nodos para los diferentes tipos de números reales
G.add_node("Números Reales")
G.add_node("Racionales")
G.add_node("Irracionales")
G.add_node("Enteros")
G.add_node("Naturales")
G.add_node("Fraccionarios")
G.add_node("Decimales")

# Agregar relaciones entre los nodos
G.add_edges_from([("Números Reales", "Racionales"),
                  ("Números Reales", "Irracionales"),
                  ("Racionales", "Enteros"),
                  ("Racionales", "Fraccionarios"),
                  ("Racionales", "Decimales"),
                  ("Enteros", "Naturales")])

# Configuración de la posición de los nodos
pos = nx.spring_layout(G, seed=42)

# Dibujar el gráfico
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, font_weight='bold', arrows=False)
plt.title("Mapa Conceptual de Clasificación de Números Reales")
plt.show()
```

# Potenciación en $\mathbb{R}$

$a^{x}$: potencia 

$a$: la base  

$x$ es el exponente. 

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize = (16, 8))

# Crear un gráfico dirigido
G = nx.DiGraph()

# Agregar el nodo principal: Potenciación en Números Reales
G.add_node("Potenciación en Números Reales")

# Agregar conceptos subordinados
G.add_node("Potenciación de Exponente Entero Positivo")
G.add_node("Potenciación de Exponente Negativo")
G.add_node("Potenciación de Exponente Fraccionario")

# Agregar relaciones entre los nodos
G.add_edges_from([("Potenciación en Números Reales", "Potenciación de Exponente\n Entero Positivo"),
                  ("Potenciación en Números Reales", "Potenciación de Exponente Negativo"),
                  ("Potenciación en Números Reales", "Potenciación de Exponente Fraccionario")])

# Configuración de la posición de los nodos
pos = nx.spring_layout(G, seed=42)

# Dibujar el gráfico
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightgreen', font_size=10, font_weight='bold', arrows=False)
plt.title("Mapa Conceptual de Potenciación en Números Reales")
plt.show()
```

# Asignación de tarea de entrenamiento de preparación del parcial



```python
import pandas as pd
import numpy as np
import random
import requests

# Descargar la lista de estudiantes y ejercicios
students_url = "https://raw.githubusercontent.com/marco-canas/groups_list/main/calculo_diferencial.csv"
exercises = list(range(1, 84))  # Lista de ejercicios del 1 al 83

# Cargar la lista de estudiantes
students_data = pd.read_csv(students_url)

# Crear una lista para almacenar las asignaciones
assignments = []

# Asignar ejercicios aleatorios a cada estudiante
for index, student in students_data.iterrows():
    random_exercises = random.sample(exercises, 5)  # Seleccionar 5 ejercicios aleatorios sin repetición
    assignments.append([students_data["Nombre"][index], *random_exercises])  # Agregar asignación a la lista

# Crear un DataFrame con las asignaciones
columns = ["Student", "Exercise 1", "Exercise 2", "Exercise 3", "Exercise 4", "Exercise 5"]
assignment_df = pd.DataFrame(assignments, columns=columns)

# Guardar el DataFrame en un archivo CSV
assignment_csv = "tarea_seccion_01_purcell.csv"
assignment_df.to_csv(assignment_csv, index=False)

print("Asignaciones guardadas en", assignment_csv)
```

2. Realizar la resta de fracciones

$$ \frac{2}{3} \left( \frac{2}{5 - \frac{1}{3}} \right)^{2} = \frac{2}{3} \left( \frac{2}{\frac{5}{1} - \frac{1}{3}} \right)^{2} $$

$$ \frac{2}{3} \left( \frac{2}{\frac{5}{1} - \frac{1}{3}} \right)^{2} = \frac{2}{3} \left( \frac{2}{ \frac{15 - 1}{3} } \right)^{2} $$

$$ \frac{2}{3} \left( \frac{2}{ \frac{15 - 1}{3} } \right)^{2} = \frac{2}{3} \left( \frac{2}{ \frac{14}{3} } \right)^{2} $$

$$ \frac{2}{3} \left( \frac{2}{ \frac{14}{3} } \right)^{2} = \frac{2}{3} \left( \frac{\frac{2}{1}}{ \frac{14}{3} } \right)^{2} $$

$$ \frac{2}{3} \left( \frac{\frac{2}{1}}{ \frac{14}{3} } \right)^{2}  = \frac{2}{3} \left( \frac{6}{14} \right)^{2}  $$
