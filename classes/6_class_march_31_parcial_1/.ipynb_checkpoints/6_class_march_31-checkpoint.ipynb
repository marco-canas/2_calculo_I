Flipped Learning o aprendizaje invertido es un enfoque pedagógico en el que la instrucción directa se realiza fuera del aula y se utiliza el tiempo de clase para llevar a cabo actividades que impliquen el desarrollo de procesos cognitivos de mayor complejidad, en las que son necesarias la ayuda y la experiencia del docente.

# Sobre las pruebas internacionales PISA  

## Qué evalúa [PISA](https://www.oecd.org/pisa/39730818.pdf)  

A diferencia de otros exámenes que se han utilizado en el pasado, PISA está diseñado para conocer las competencias, o, dicho en otros términos, las habilidades, la pericia y las aptitudes de los estudiantes para 
* analizar y resolver problemas, 
* para manejar información y 
* para enfrentar situaciones que se les presentarán en la vida adulta y que requerirán de tales habilidades.
PISA se concentra en la evaluación de tres áreas: 
* competencia lectora, 
* competencia matemática y 
* competencia científica.

```python
# Un gráfico de líneas. 

import numpy as np # traer la herramienta de funciones matemáticas 
import matplotlib.pyplot as plt 

plt.title ("Temperatura en Caucasia en horas de la tarde y noche") # dar título al gráfico

plt.xlabel ("hora $t$")
plt.ylabel("temperatura $T$")

t=["12:00", "13:00", "14:00", "15:00", "16:00", "17:00", "18:00", "19:00", "20:00"]
T=[32, 34, 35, 35, 34, 33, 32, 30, 29]
plt.plot(t, T, color = "blue")
plt.grid()
plt.scatter (t,T)
plt.show()
```

* A las 12 y a las 6PM la temperatura en Caucasia es la misma. 
* A las 1 M y a las 4 PM la temperatura en Caucasia es la misma. 
* A las 2 PM y a las 3 PM la temperatura en Caucasia es la misma y es la máxima del día.
