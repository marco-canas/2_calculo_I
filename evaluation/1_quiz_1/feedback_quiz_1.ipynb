```python
from sympy import * 
x = var('x')

funcion_racional = (x-4)/(x**2-16) # 


import matplotlib.pyplot as plt
import numpy as np

f = lambdify(x, funcion_racional, "numpy")
f(3.99999)
```

### Del análisis puntual deducimos que a medida que x tiende a 4 por izquierda, los valores de la función racional tienden a 0.125 = 1/8

```python
x_puntos = np.linspace(3.8, 4.2, 10)
y_valores = f(x_puntos)
plt.plot(x_puntos, y_valores)       # para trazar líneas poligonales 
plt.axhline(y=0, color='k', lw=0.5) # eje horizontal 
plt.axvline(x=4, color='k', lw=0.5) # recta vertical 
plt.ylim([-0.1, 0.3])
plt.title(r"Gráfico de la función racional $\frac{x-4}{x^{2} - 16}$")
plt.xlabel("x")
plt.ylabel("y")
plt.yticks(np.arange(0.1, 0.2, 0.025))
plt.grid() # para adicionarl la cuadrícula
plt.savefig('comportamiento_de_la_funcion_en_4.png')

plt.show()
```

### Interpretaciones o lecturas al gráfico anterior

* la tendencia de la función racional dada es hacia un valor menor a $0.150$.
* A pesar de que esta función no está definida en $x = 4$, si tiene una tendencia definida en este punto y es hacia un valor menor a 0.150. 

## Es necesario una gráfico de dispesión animado 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook
epsilon = 0.001
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = ax.plot([], [], 'ro')

def init(): # definió el rectángulo de visualización
    ax.set_xlim(3, 4.1)
    ax.set_ylim(0.09, 0.1)
    ax.set_yticks(np.arange(0.1, 0.2, 0.025))
    ax.grid(alpha = 0.4)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(f(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(3, 4, 50),
                    init_func=init, blit=True)
plt.show()
```

## Observaciones obtenidas del proceso de animación por izquierda de un límite

* Los valores de la función tienden a decrecer hacia $0.125$ a medida que los puntos $x$ se van acercando a 4 por la izquierda. 

# Obtención del límite utilizando el teorema de las funciones casi iguales en todas partes.

```python
funcion_casi_igual = simplify(funcion_racional)
funciona_casi_igual
```

### Conclusión

La función casi igual a la función racional dada es 

$$ \frac{1}{x+4} $$

y esta función evaluada en $x = 4$ es igual a $1/8 = 0.125$.

La ventaja de obtener la función casi igual es poder hallar el límite solo sustituyendo o evaluando en el punto dado.
