4.  Una larga hoja de metal, de 16 pulgadas de ancho, se dobla
hacia arriba en ambos lados para formar un canalón horizontal con
lados verticales. ¿Cuántas pulgadas de cada lado deben doblarse hacia arriba para maximizar la capacidad de carga? (40)


Respuesta:

base del canalón será de 16 - 2x pulgadas (debido a que se doblan ambos lados). La altura del canalón será x pulgadas.

La capacidad de carga de un canalón se puede modelar como el producto del ancho, el alto y la longitud del canalón. En este caso, dado que la longitud no se especifica, la trataremos como una constante k (en pulgadas).

La capacidad de carga C(x) se calcula como:
$C(x) = (16 - 2x) * x * k$

Para maximizar la capacidad de carga, necesitamos encontrar el valor de x que maximice la función C(x).

Podemos derivar C(x) con respecto a x y encontrar el punto crítico donde la derivada sea igual a cero o no esté definida.

$C'(x) = -4x + 16k$

Igualamos C'(x) a cero:
$-4x + 16k = 0$
$4x = 16k$
$x = 4k$

Dado que x representa las pulgadas dobladas hacia arriba en cada lado, x = 4k implica que se deben doblar 4k pulgadas hacia arriba en cada lado.

##Conclusión:

para maximizar la capacidad de carga del canalón, se deben doblar 4k pulgadas hacia arriba en cada lado, donde k es una constante y representa la longitud del canalón en pulgadas.

longitud del tablón es x pies. La altura de la barda es de 8 pies y la distancia entre la barda y el muro es de 1 pie.

La longitud del tablón, que llamaremos L, se puede encontrar utilizando el teorema de Pitágoras, que establece que la hipotenusa al cuadrado es igual a la suma de los cuadrados de los catetos en un triángulo rectángulo. En este caso, la hipotenusa es la longitud del tablón y los catetos son la altura de la barda y la distancia entre la barda y el muro.

Entonces, tenemos:
$L^2 = x^2 + 8^2$



consideremos la función $L(x) = x^2 + 8^2$ y busquemos el mínimo de esta función.

Calculamos la derivada de L(x) con respecto a x:
$L'(x) = 2x$

Para encontrar el mínimo, igualamos la derivada a cero y resolvemos la ecuación:


$2x = 0$
$x = 0$

$x = 0$ no es un valor válido, por lo tanto, debemos verificar los puntos críticos en los límites del dominio.

El dominio del problema está definido por la distancia entre la barda y el muro, que es de 1 pie. Por lo tanto, el dominio de x está restringido a [0, 1].

Evaluamos la función L(x) en los extremos del dominio:
$L(0) = 0^2 + 8^2 = 64$
$L(1) = 1^2 + 8^2 = 65$

Vemos que $L(0) = 64 y L(1) = 65$.

Por lo tanto, la longitud mínima del tablón que puede pasar por encima de la barda y apuntalar el muro es de 64 pies, que ocurre cuando x = 0.

ancho de la página es x pulgadas y la altura de la página es y pulgadas.

ancho y la altura:
$A = x * y$

* Entonces, el área de impresión:
$A = (x - 4) * (y - 5)$

* Dado que el área de impresión es de 27 pulgadas cuadradas:
$(x - 4) * (y - 5) = 27$

encontrar las dimensiones de página que minimicen el área, necesitamos minimizar la función A sujeta a la restricción dada.

* método de sustitución:

$A = (x - 4) * \left(\frac{27}{x - 4} + 5\right)$

$A = 27 + 5(x - 4)$

6.  Se lanza una pelota directamente hacia arriba desde una torre
de 448 pies de altura, a una velocidad inicial de 48 pies por segundo.
¿En cuántos segundos chocará con el piso y a qué velocidad? Suponga que g = 32 pies por segundo por segundo y no tome en cuenta la
resistencia del aire. (75)

La posición de la pelota en función del tiempo se puede describir con la ecuación:

$h(t) = h0 + v0t - (1/2)gt^2$

Donde:

* h(t) es la posición en función del tiempo t.
* h0 es la altura inicial.
* v0 es la velocidad inicial.
* g es la aceleración debido a la gravedad.
* t es el tiempo.


valor de t cuando la posición h(t) es igual a cero.

$0 = h0 + v0t - (1/2)gt^2$

Rearreglando la ecuación:

$(1/2)gt^2 - v0t - h0 = 0$

Esta es una ecuación cuadrática en t. Para encontrar el valor de t, podemos utilizar la fórmula cuadrática.

La fórmula cuadrática es:

$t = (-b ± sqrt(b^2 - 4ac)) / (2a)$

Donde:

* $a = (1/2)g$
* $b = -v0$
* $c = -h0$


Sustituyendo los valores en la fórmula cuadrática, obtenemos:

$t = (-(-v0) ± sqrt((-v0)^2 - 4((1/2)g)(-h0))) / (2((1/2)g))$

Simplificando la expresión:

$t = (v0 ± sqrt(v0^2 + 2gh0)) / g$



v0 = 48 pies/s
g = 32 pies/s^2
h0 = 448 pies

Sustituyendo los valores, obtenemos:

$t = (48 ± sqrt(48^2 + 2(32)(448))) / 32$

Simplificando la expresión:

* $t = (48 ± sqrt(2304 + 28672)) / 32$

* $t = (48 ± sqrt(30976)) / 32$

* $t = (48 ± 176) / 32$

Tenemos una solucion posible:

* $t1 = (48 + 176) / 32 = 6.875 segundos$


Por lo tanto, la pelota chocará con el piso aproximadamente a los 6.875 segundos.

Para determinar la velocidad con la que chocará, podemos utilizar la derivada de la función de posición.

La derivada de la función de posición h(t) con respecto al tiempo t nos da la función de velocidad v(t):

$v(t) = h'(t) = v0 - gt$

Sustituyendo los valores:

$v(t) = 48 - 32t$

Evaluamos la velocidad en t = 6.875 segundos:

$v(6.875) = 48 - 32(6.875) ≈ -36.5 ft/s$

Por lo tanto, la pelota chocará con el piso a una velocidad aproximada de -36.5 pies por segundo. El signo negativo indica que la velocidad es hacia abajo, en dirección opuesta al lanzamiento inicial.

7. Un abrevadero metálico con extremos semicirculares iguales,
sin cubierta superior, debe tener una capacidad de 128p pies cúbicos
(véase la figura 1). Determine su radio r y longitud h, si el abrevadero debe requerir la menor cantidad de material para su construcció. (43)



La superficie S de un abrevadero semicircular sin cubierta superior está dada por la fórmula:

$S = πr^2 + 2πrh$

Queremos minimizar la superficie S, por lo tanto, tomaremos la derivada de S con respecto a r y haremos igual a cero para encontrar los puntos críticos.

Para derivar S respecto a r, consideraremos h como una constante:

$dS/dr = 2πr + 2πh(dr/dr) = 2πr$

Igualamos a cero:

$2πr = 0$

Esto nos da r = 0 como uno de los puntos críticos, pero esto no tiene sentido en el contexto del problema, ya que el radio debe ser positivo.

Ahora derivamos S respecto a h, considerando r como una constante:

$dS/dh = 2πr$


Por lo tanto, no tenemos puntos críticos en los cuales la superficie pueda alcanzar un mínimo.

Sin embargo, dado que la capacidad del abrevadero está fija en 128p pies cúbicos, podemos utilizar esta información para encontrar el valor de r y h.

El volumen V del abrevadero está dado por:

$V = (1/2)πr^2h$

Sustituyendo el valor de $V = 128p$ y resolviendo para h:

$h = (2V) / (πr^2) = (2 * 128p) / (πr^2) = 256p / (πr^2)$

La superficie S en términos de r y h es:

$S = πr^2 + 2πrh = πr^2 + 2πr * (256p / (πr^2)) = πr^2 + 512p / r$

Para minimizar la superficie, podemos derivar S con respecto a r y encontrar los puntos críticos:

$dS/dr = 2πr - 512p / r^2 = 0$

Multiplicando por $r^2$:

$2πr^3 - 512p = 0$

Resolviendo para r:

$r^3 = 256p / π$

$r = (256p / π)^(1/3)$

Sustituyendo este valor de r en la ecuación de h:

$h = 256p / (π * ((256p / π)^(1/3))^2) = 256p / (256p / π)^(2/3) = (256p)^(1/3)$

Por lo tanto, el radio r que minimiza la superficie del abrevadero es $(256p / π)^(1/3) $y la longitud h correspondiente es $(256p)^(1/3)$.
