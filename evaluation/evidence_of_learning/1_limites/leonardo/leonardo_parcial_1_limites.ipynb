Respuesta 3: Podemos factorizar el denominador usando la diferencia de cuadrados, de la siguiente manera:



$f(x) = (x-1)/(x^2-1)$

$lim_{x->1} f(x) = lim_{x->1} (x-1)/(x^2-1)$

$lim_{x->1} f(x) = lim_{x->1} (x-1)/[(x-1)(x+1)]
                 = lim_{x->1} 1/(x+1)$

$lim_{x->1} f(x) = 1/(1+1) = 1/2$







```python
from sympy import *

# Definir la variable x
x = symbols('x')

# Definir la función
f = (x-1)/(x**2-1)

# Calcular el límite cuando x se acerca a 1
lim = limit(f, x, 1)

# Imprimir el resultado
print("El límite cuando x se acerca a 1 es:", lim)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definir la función
def f(x):
    return (x-1)/(x**2-1)

# Crear un arreglo de valores de x
x = np.linspace(-10, 10, 500)

# Calcular los valores de y correspondientes
y = f(x)

# Crear la gráfica
plt.plot(x, y)

# Establecer el título y los nombres de los ejes
plt.title('Gráfica de la función (x-1)/(x^2-1)')
plt.xlabel('x')
plt.ylabel('y')

# Mostrar la gráfica
plt.show()
```

La conclusión del límite de la función $\lim_{x \to 1} \frac{x-1}{x^2-1}$ es que cuando $x$ se acerca a $1$, el valor de la función se acerca un $\frac{1}{2}$. Esto significa que la función tiene un límite bien definido en $x=1$ y que podemos calcular el valor de la función en ese punto a partir de su límite. Respuesta 4: Para solucionar este lìmite, se realizara lo siguiente:

Primero, podemos factorizar la diferencia de cuadrados en el denominador:

$$f(x) = (\sqrt{x} - 1)/(x - 1) * (\sqrt{x} + 1)/(\sqrt{x} + 1)$$
 $$    = (\sqrt{x} - 1)(\sqrt{x} + 1)/(x - 1)(\sqrt{x} + 1)$$
  $$   = (x - 1)/(x - 1)(\sqrt{x} + 1)$$
   $$  = 1/(\sqrt{x} + 1)$$

   A continuación, podemos sustituir x = 1 en la función simplificada:

$$lim_{x -> 1} f(x) = lim_{x -> 1} 1/(\sqrt{x} + 1) = 1/2$$

Por lo tanto, el límite de la función $f(x) = (\sqrt{x} - 1)/(x - 1) $ cuando x se acerca a 1 es igual a 1/2.

```python
# De manera gràfica serìa lo siguiente:

import matplotlib.pyplot as plt
import numpy as np

# Definir la función f(x)
def f(x):
    return (np.sqrt(x) - 1) / (x - 1)

# Crear un arreglo de valores x cercanos a 1
x = np.linspace(0.9, 1.1, 1000)

# Graficar la función f(x)
plt.plot(x, f(x))

# Agregar una línea vertical en x=1
plt.axvline(x=1, linestyle='--', color='gray')

# Agregar un título y etiquetas de eje
plt.title(r'Límite de $\frac{\sqrt{x}-1}{x-1}$ cuando $x$ se acerca a 1')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')

# Mostrar la gráfica
plt.show()
```

$$\lim_{x\to 1}\left(\frac{\frac{1}{x}-\frac{1}{1}}{x-1}\right)$$

$$
\begin{aligned}
\lim_{x\to 1}\left(\frac{\frac{1}{x}-\frac{1}{1}}{x-1}\right) &= \lim_{x\to 1}\left(\frac{\frac{1- x}{x}}{x-1}\right) \\
&= \lim_{x\to 1}\left(\frac{1-x}{x(x-1)}\right) \\
&= \lim_{x\to 1}\left(\frac{-(x-1)}{x(x-1)}\right) \\
&= \lim_{x\to 1}\left(\frac{-1}{x}\right) \\
&= -1
\end{aligned}
$$
