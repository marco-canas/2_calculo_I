## Solución

a. El número de Euler "e" se define matemáticamente como el límite de la función exponencial "e^x" cuando "x" tiende a infinito, es decir:

$$ e = \lim_{x \to \infty} \left(1 + \frac{1}{x} \right)^x $$

Esta definición implica que el número de Euler "e" es el valor único tal que la función exponencial "$e^{x}$" crece a una tasa igual a su propio valor en cualquier punto "$x$". Es un número irracional y tiene una aproximación decimal de aproximadamente

$$ 2.718281828459045...$$

El número de Euler "e" tiene muchas propiedades interesantes y se encuentra involucrado en una amplia variedad de áreas en matemáticas, como el cálculo diferencial e integral, la teoría de números, la probabilidad, la estadística, entre otras. También es fundamental en el estudio de funciones exponenciales y logarítmicas.

Una de las formas más comunes de utilizar el número de Euler "e" es en la función exponencial:

$$ e^{x} = 1 + x + (x^2)/2! + (x^3)/3! + (x^4)/4! + ... $$

Donde "x" es un número real, y "n!" representa el factorial de "n" (el producto de todos los enteros positivos desde 1 hasta "n"). Esta serie infinita es convergente para cualquier valor de "x", lo que significa que la suma de todos los términos de la serie converge a un valor finito para cualquier valor de "x".

El número de Euler "e" es una constante fundamental en matemáticas y se utiliza ampliamente en diversas aplicaciones, desde cálculos científicos y financieros hasta problemas teóricos y abstractos en matemáticas puras.

b. Estimemos numéricamente el primer límite

 $$ \lim_{h \to 0} \frac{2.7^{h}-1}{h}  $$


```python
import numpy as np

H_derecha = np.array([1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001, 0.0000001])
H_izquierda = np.array([-1, -0.1, -0.01, -0.001, -0.0001, -0.00001, -0.000001, -0.0000001])

def f(h):
  return (2.7**h - 1)/h

cocientes_derecha = f(H_derecha)
cocientes_izquierda = f(H_izquierda)

print(cocientes_derecha)
print(cocientes_izquierda)
```

### Conclusión

$$ \lim_{h \to 0} \frac{2.7^{h}-1}{h} \approx 0.99325 $$

```python
import numpy as np

H_derecha = np.array([1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001, 0.0000001])
H_izquierda = np.array([-1, -0.1, -0.01, -0.001, -0.0001, -0.00001, -0.000001, -0.0000001])

def g(h):
  return (2.8**h - 1)/h

cocientes_derecha = g(H_derecha)
cocientes_izquierda = g(H_izquierda)

print(cocientes_derecha)
print(cocientes_izquierda)
```

### Conclusión
