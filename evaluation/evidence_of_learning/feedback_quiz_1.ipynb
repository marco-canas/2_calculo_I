$$\lim _{x\to 4}\left(\frac{x-4}{x^2-16}\right)$$

## Observaciones del profesor al trabajo de los estudiantes

* Debo enfatizar en el teorema de las funciones casi iguales en todas partes y en la necesidad de simplificar una fracción para encontrar una función simple en donde se pueda determinar el límite por sustitución o evaluación. 

```python
from calculo_limite_tabular_grafico import tabla_izquierda, tabla_derecha
```

```python
def f(x):
    return (x-4)/(x**2-16)
```

```python
tabla_izquierda(f, c = 4)
```

```python
from sympy import * 
x = var('x')

expr = (x-4)/(x**2-16) # 


import matplotlib.pyplot as plt
import numpy as np

f = lambdify(x, expr, "numpy")
f(3.999)
```

```python
x_vals = np.linspace(3.5, 4.5, 1000)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals)
plt.axhline(y=0, color='k', lw=0.5)
plt.axvline(x=4, color='k', lw=0.5)
plt.ylim([-10, 10])
plt.title("Gráfico de la función")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Es necesario una gráfico de dispesión animado 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook
epsilon = 0.001
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = ax.plot([], [], 'ro')

def init(): # definió el rectángulo de visualización
    ax.set_xlim(3, 4.1)
    ax.set_ylim(0.09, 0.1)
    ax.set_yticks(np.arange(0.1, 0.2, 0.025))
    ax.grid(alpha = 0.4)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(f(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(3, 4, 50),
                    init_func=init, blit=True)
plt.show()
```
