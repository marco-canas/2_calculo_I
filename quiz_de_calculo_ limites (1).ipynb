# DERERMINAR EL SIGUIENTE LIMITE:



$$ \lim_{x \to 7} \frac{x - 7}{x^{2} - 49} $$

Empezamos a determinar el limite analiticamente
y para ello ensayamos primero sustituyendo.

$$ \lim_{x \to 3} \frac{x - 3}{x^{2} - 9} =  \frac{(3) - 3}{(3)^{2}- 9}=\frac{3 - 3}{9- 9}=  \frac{0 }{ 0}$$

Podemps obserbar que que se obtuvo una indeterminacion, es decir no se pudo obtener el limite sustituyendo.

Debemos buscar otra solucion

Podemos utilizar el metodo de factorizacion.

$$ \lim_{x \to 3} \frac{x - 3}{x^{2} - 9}  =  \lim_{x \to 3}  \frac{x - 3}{(x+3)(x-3)}=\frac{1}{3+ 3}=  \frac{1}{ 6}$$

luego simplificamos y obtenemos el limire exacto
